```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
contract_table = df.groupby(["Contract","Churn"])["customerID"].count().unstack()
contract_table
```

```python
contract_table["churn_rate"] = contract_table["Yes"]/(contract_table["Yes"] + contract_table["No"])
contract_table
```

```python
contract_table["churn_rate"].plot(kind="bar")

plt.title("Churn Rate by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Churn Rate")

plt.show()
```

```python
df.groupby(["tenure","Churn"])["customerID"].count()
```

```python
df["tenure_group"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 100],
    labels=["0-12", "13-24", "25-48", "48+"]
)
```

```python
df.head()
```

```python
tenure_table = df.groupby(["tenure_group","Churn"])["customerID"].count().unstack()
tenure_table
```

```python
tenure_table["tenure_churn_rate"] = tenure_table["Yes"]/(tenure_table["Yes"] + tenure_table["No"])
tenure_table
```

```python
tenure_table["tenure_churn_rate"].plot(kind="bar")

plt.title("Churn Rate by Customer Tenure")
plt.xlabel("Tenure Group")
plt.ylabel("Churn Rate")

plt.show()
```

Early-tenure customers represent the highest churn risk, indicating the need for stronger onboarding and early engagement strategies.

```python
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)

plt.title("Monthly Charges by Churn Status")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

Customers paying higher monthly charges exhibit higher churn tendencies, indicating possible pricing sensitivity or perceived value gaps in premium plans.
